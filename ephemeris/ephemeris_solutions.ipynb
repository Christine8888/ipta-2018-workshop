### Part 1: What's the difference between the different ephemerides?

The Solar System Ephemeris (SSE) is a model that describes the positions of the Sun, planets, major moons, and asteroids. In pulsar timing, we use the SSE to figure out the location of the Solar System barycenter (SSB).

Let's load in the two most recent JPL ephemerides: DE435 and DE436.

```python
ephemdir = '/opt/pulsar/share/tempo2/ephemeris/'

# This will read in the orbits contained in the .bsp file. The SSB is item 0, and planets are 1-9.
bsp435 = SPK.open(ephemdir + 'de435t.bsp')
bsp436 = SPK.open(ephemdir + 'de436t.bsp')
print(bsp435)
```

This function will read in the ephemeris file and return an array containing the orbit of a given planet.

```python
def getplanet(efile,i,jdates):
    """Return a len(jdates) x 3 numpy array of the orbit for planet i,
    following the ephemeris file numbering (3 is Earth, 10 is Sun).
    Special cases: 310 is Sun to Earth, 399 is Earth bary + nutation.
    The array is of x,y,z equatorial coordinates in km."""
    
    bsp = SPK.open(efile)
    
    if i == 399:
        return (bsp[0,3].compute(jdates).T + bsp[3,399].compute(jdates).T)
    elif i == 310:
        return (bsp[0,3].compute(jdates).T - bsp[0,10].compute(jdates).T)
    else:
        return bsp[0,i].compute(jdates).T
```

The function I have given you gives the orbit in rectangular coordinates (x,y,z), but it would be more convenient to have it in spherical coordinates. Write a function that converts it from rectangular coordinates to spherical.

```python
def xyz2sph(pos):
    r = np.sqrt(pos[:,0]**2 + pos[:,1]**2 + pos[:,2]**2)    
    lon = np.arctan2(pos[:,0],pos[:,1])
    lat = np.arcsin(pos[:,2]/np.sqrt(pos[:,0]**2 + pos[:,1]**2))
    
    return np.array([r,lon,lat]).T
```

Now use the function I gave you and the function you wrote to plot the position of Jupiter using DE436 over a time span of about 20 years.

Notice that the ephemerides use the time in Julian date, not modified Julian date.
They are related by jd = mjd + 2400000.5.

```python
mjd = np.linspace(52275,59214)

jupiter_436 = getplanet(ephemdir + 'de436t.bsp', 5, mjd+2400000.5)
jupiter_436_Sph = xyz2sph(jupiter_436)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(6,8))

axs[0].plot(mjd, jupiter_436_Sph[:,0])
axs[1].plot(mjd, jupiter_436_Sph[:,1])
axs[2].plot(mjd, jupiter_436_Sph[:,2])

for ax in axs:
    ax.set_xlabel('MJD')
    
axs[0].set_ylabel('Radius (km)')
axs[1].set_ylabel('Longitude')
axs[2].set_ylabel('Latitude')
    
plt.tight_layout()
```

But what we really care about is the difference between two different ephemerides.

Do the same thing as above, but this time use DE435 as the ephemeris. Plot the difference in the orbit of Jupiter between DE435 and DE436.

```python
jupiter_435 = getplanet(ephemdir + 'de435t.bsp', 5, mjd+2400000.5)
jupiter_435_Sph = xyz2sph(jupiter_435)

fig, axs = plt.subplots(3, 1, figsize=(6,8))

axs[0].plot(mjd, jupiter_435_Sph[:,0]-jupiter_436_Sph[:,0])
axs[1].plot(mjd, jupiter_435_Sph[:,1]-jupiter_436_Sph[:,1])
axs[2].plot(mjd, jupiter_435_Sph[:,2]-jupiter_436_Sph[:,2])

for ax in axs:
    ax.set_xlabel('MJD')
    
axs[0].set_ylabel('Change in Radius (km)')
axs[1].set_ylabel('Change in Longitude')
axs[2].set_ylabel('Change in Latitude')
    
plt.tight_layout()
```

Now let's look at how the position of the Earth with respect to the SSB varies with ephemeris.

```python
mjd = np.linspace(52275,59214)

earth_435 = getplanet(ephemdir + 'de435t.bsp', 3, mjd+2400000.5)
earth_436 = getplanet(ephemdir + 'de436t.bsp', 3, mjd+2400000.5)

earth_435_Sph = xyz2sph(earth_435)
earth_436_Sph = xyz2sph(earth_436)

fig, axs = plt.subplots(3, 1, figsize=(6,8))

axs[0].plot(mjd, earth_435_Sph[:,0]-earth_436_Sph[:,0])
axs[1].plot(mjd, earth_435_Sph[:,1]-earth_436_Sph[:,1])
axs[2].plot(mjd, earth_435_Sph[:,2]-earth_436_Sph[:,2])

for ax in axs:
    ax.set_xlabel('MJD')
    
axs[0].set_ylabel('Change in Radius (km)')
axs[1].set_ylabel('Change in Longitude')
axs[2].set_ylabel('Change in Latitude')
    
plt.tight_layout()
```

Finally, let's convert the change in orbital radius from km to s.

```python
c = 3e5

fig, ax = plt.subplots(1, 1, figsize=(6,3))

ax.plot(mjd, (earth_435_Sph[:,0]-earth_436_Sph[:,0])/c)

ax.set_xlabel('MJD')   
ax.set_ylabel('Change in Radius (s)')
    
plt.tight_layout()
```

### Part 2: Roemer delay differences in pulsars

Now let's take a look at the Roemer delay for an actual pulsar under different ephemerides.

Load up the par and tim files for a pulsar (I have given you J1909-3744 in the partim directory). You can specify the ephemeris with the option `ephem='EPHEM'`. Use all four of the most recent ephemerides (DE421, DE430, DE435, and DE436).

```python
import libstempo as LT

parfile = 'partim/J1909-3744_NANOGrav_11yv0.gls.par'
timfile = 'partim/J1909-3744_NANOGrav_11yv0.tim'

psr_DE421 = LT.tempopulsar(parfile=parfile, timfile=timfile, ephem='DE421')
psr_DE430 = LT.tempopulsar(parfile=parfile, timfile=timfile, ephem='DE430')
psr_DE435 = LT.tempopulsar(parfile=parfile, timfile=timfile, ephem='DE435')
psr_DE436 = LT.tempopulsar(parfile=parfile, timfile=timfile, ephem='DE436')
```

You can get the Roemer delay for a pulsar with `psr.roemer`. It will return an array of the Roemer delay in seconds at each observation. First, plot the Roemer delay for this pulsar using DE436.

```python
plt.plot(psr_DE436.toas(), psr_DE436.roemer, label='Roemer delay, DE436')
plt.legend(loc=2)

plt.xlabel('MJD')
plt.ylabel('Time (s)')
```

Next, plot the differences between different ephemerides.

```python
plt.plot(psr_DE436.toas(), psr_DE430.roemer-psr_DE436.roemer, label='DE430-DE436')
plt.plot(psr_DE436.toas(), psr_DE435.roemer-psr_DE436.roemer, label='DE435-DE436')
plt.plot(psr_DE436.toas(), psr_DE421.roemer-psr_DE436.roemer, label='DE421-DE436')
plt.legend()
```

Notice that these plots look different from Michele's. That's because the plots he generated perform quadratic subtraction to simulate fitting the timing model. We are not going that -- we are just plotting the value of the Roemer delay.
